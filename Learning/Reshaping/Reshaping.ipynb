

>**RESHAPING**



bentuk data untuk memudahkan proses analisis dengan 2 teknik yaitu pivot data dan melting data
- Pivot : perubahan bentuk data dari bentuk melebar ke bawah menjadi melebar ke samping
- Melting : perubahan bentuk data dari bentuk melebar menjadi memanjang



> **PERSIAPAN DATASET**



dataset : https://www.kaggle.com/datasets/datatechexplorer/samsung-mobile-sales-dataset

```python
import pandas as pd

data=pd.read_csv("Samsung sales.csv", sep=";")
data
```


>**PIVOTING DATAFRAME**

3 parameter yang harus diikutsertakan yaitu
- index : kolom yang akan dijadikan index
- column : kolom yang akan dijadikan kolom
- values : kolom yang akan dijadikan values

**Pivot Data**

Jumlah Produk Terjual (Units Sold) Berdasarkan Tahun (Year) dan Model Produk (Product Model)

```python
data_pivot=data.pivot_table(
    index="Year",
    columns="Product Model",
    values="Units Sold",
    aggfunc="sum").applymap(lambda x:f"{x:,}") #menambah separator ribuan
data_pivot
```

**Pivot Data**

Jumlah Produk Terjual (Units Sold) Berdasarkan Tahun (Year) dan Model Produk (Product Model) dengan tambahan Baris Total dan Kolom Total


```python
data_pivot=data.pivot_table(
    index="Year",
    columns="Product Model",
    values="Units Sold",
    aggfunc="sum",
    margins=True,
    margins_name="Total",
    fill_value=0).applymap(lambda x: f"{x:,}") #menambah separator ribuan
data_pivot
```

**Pivot Data Dengan Filter**

Jumlah Produk Terjual Berdasarkan Tahun dan Model Produk untuk Periode 2023 dan 2024

```python
data_pivot.loc[[2023, 2024]]
```

**Pivot Data**

Jumlah Produk Terjual (Units Sold) Berdasarkan Tahun (Year) dan Wilayah (Region)

```python
data_pivot1=data.pivot_table(
    index="Year",
    columns="Region",
    values="Units Sold",
    aggfunc="sum").applymap(lambda x: f"{x:,}") #menambah separator ribuan
data_pivot1
```

**Pivot Data**

Jumlah Produk Terjual (Units Sold) Berdasarkan Tahun (Year) dan Wilayah (Region) dengan tambahan Baris Total dan Kolom Total

```python
data_pivot1=data.pivot_table(
    index="Year",
    columns="Region",
    values="Units Sold",
    aggfunc="sum",
    margins=True,
    margins_name="Total").applymap(lambda x: f"{x:,}") #menambahkan separator ribuan
data_pivot1
```

**Pivot Data Dengan Filter**

Jumlah Produk Terjual Berdasarkan Tahun dan Wilayah untuk Periode 2023, 2024 dengan Wilayah Europe dan North Amerika

```python
data_pivot1.loc[[2023,2024], ["Europe","North America"]]
```



> **MELTING DATAFRAME**

Menggunakan *method* melt()

Parameter yang digunakan yaitu
- id_vars : kolom yang akan dijadikan id
- value_vars : kolom yang akan di melting
- var_name : kolom untuk hasil melting
- value_name : kolom yang dijadikan values



```python
#melting data menggunakan data_pivot1
melt_class=pd.melt(
    data_pivot1.reset_index(),
    id_vars="Year",
    value_vars=["Asia-Pacific","Europe","Latin America", "Middle East & Africa","North America"],
    var_name="Region",
    value_name="Units Sold"
)
melt_class
```
