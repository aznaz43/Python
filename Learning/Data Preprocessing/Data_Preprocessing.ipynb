**IDENTIFIKASI MISSING VALUE**

```python
# mendefinsiikan format missing value
missing_value=["N.A", "na", "n.a.", "n/a", "-"," "]

# menmabhkan parameter na_values untuk memformat missing_value
data = pd.read_csv(dataset, na_values=missing_value)
data.info()
```

**MEMERIKSA MISSING VALUE DATAFRAME**

```python
# memeriksa missing value kseseluruhan
data.isna().values.any()

# memeriksa missing values setiap kolom
data.isna().any()
```

**MENGHITUNG JUMLAH MISSING VALUE**

```python
# memeriksa missing value tiap kolom
data.isna().sum()
```

**MENGHAPUS BARIS YANG MENGANDUNG MISSING VALUE**

```python
# menghapus baris yang mengandung missing value pada kolom Stock
data.dropna(subset=["Stock"], axis=0, inplace=True)

# mereset index karena ada data terhapus
data.reset_index(drop=True, inplace=True)

data
```

**MENGISI MISSING VALUE DENGAN NILAI TERTENTU**

```python
# mengisi missing value pada kolom Category dengan Other
data["Category"].fillna("Other", inplace=True)

#mengisi missing value pada kolom discount dengan 0
data["Discount"].fillna(0,inplace=True)
data
```

**MENGISI MISSING VALUE DENGAN MEAN/MEDIAN**

```python
# mengisi missing value pada kolom Price dengan mean
data["Price"].fillna(data["Price"].mean(), inplace=True)

# mengisi missing value pada kolom Rating dengan median
data["Rating"].fillna(data["Rating"].median(), inplace=True)

data
```

```python
data.isna().sum()
```
