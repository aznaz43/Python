```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data=pd.read_csv("employee_data.csv")
data
```

**Memeriksa Missing Value**

```python
data.isna().sum()
```

```python
#menghapus kolom ExitDate dan TerminationDescription

data.drop(columns=["ExitDate","TerminationDescription"])
```

```python
# menghitung jumlah karyawan berdasarkan gender
gender_count=data["GenderCode"].value_counts()
gender_count
```

```python
# membuat pie chart dengan label persentase
plt.figure(figsize=(6,6))
mycolors=["skyblue", "pink"]
mylabels=["Laki-Laki","Perempuan"]

plt.pie(
    gender_count,
    labels=mylabels,
    autopct="%1.1f%%",
    startangle=90, #potongan pertama mulai arah atas (jam 12)
    colors=mycolors
    )

plt.title("Distribusi Karyawan Berdasarkan Jenis Kelamin")
plt.show()
```

```python
# membuat pie chart dengan label angka
plt.figure(figsize=(6,6))
mycolors=["skyblue", "pink"]
mylabels=["Laki-Laki","Perempuan"]

plt.pie(
    gender_count,
    labels=mylabels,
    autopct=lambda pct: "{:d}".format(int(round(pct/100.*sum(gender_count)))),
    startangle=90,
    colors=mycolors
    )

plt.title("Distribusi Karyawan Berdasarkan Jenis Kelamin")
plt.show()
```

```python
# membuat pie chart dengan menambahkan legend
plt.figure(figsize=(6,6))
mycolors=["skyblue", "pink"]
mylabels=["Laki-Laki","Perempuan"]

plt.pie(
    gender_count,
    autopct=lambda pct: "{:d}".format(int(round(pct/100.*sum(gender_count)))),
    startangle=90,
    colors=mycolors
    )

plt.title("Distribusi Karyawan Berdasarkan Jenis Kelamin")
plt.legend(mylabels, title="Gender")
plt.show()
```

```python
# menghitung jumlah karyawan berdasarkan tipe klasifikasi karyawan
type_count=data['EmployeeClassificationType'].value_counts()
type_count
```

```python
# membuat pie chart distribusi karyawan berdasarkan status
plt.figure(figsize=(6,6))
colors=["skyblue","lightgreen", "orange"]

plt.pie(
    type_count,
    labels=type_count.index,
    autopct="%1.1f%%",
    startangle=90,
    colors=colors
)

plt.title("Distribusi Karyawan Berdasarkan Status")
plt.show()
```

```python
plt.figure(figsize=(6,6))
colors=["skyblue","lightgreen", "orange"]

plt.pie(
    type_count,
    autopct="%1.1f%%",
    startangle=90,
    colors=colors,
)

plt.title("Distribusi Karyawan Berdasarkan Status")
plt.legend(type_count.index, title="Status Karyawan")
plt.show()
```

```python
plt.figure(figsize=(6,6))
colors=["skyblue","lightgreen", "orange"]

plt.pie(
    type_count,
    autopct="%1.1f%%",
    startangle=90,
    colors=colors,
    explode=[0.1,0.1,0.1],
    shadow=True
)

plt.title("Distribusi Karyawan Berdasarkan Status")
plt.legend(type_count.index, title="Status Karyawan")
plt.show()
```
