

> **PERSIAPAN DATASET**



dataset yang digunakan adalah dataset penjualan samsung mobile.

https://www.kaggle.com/datasets/datatechexplorer/samsung-mobile-sales-dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data=pd.read_csv("Samsung sales.csv", sep=";")
data
```

**MEMERIKSA MISSING VALUE**

```python
# memeriksa missing value
data.isna().sum()
```

Berdasarkan hasil didatas, tidak ditemukan missing value pada data.

```python
data.info()
```



> **VISUALISASI DATA DENGAN BAR PLOT**



```python
# membuat datafrae total data untuk masing-masing product model
group=data.groupby("Product Model").size()
data_total=pd.DataFrame(group).reset_index()
data_total.columns=["Product Model","Total"]
data_total
```

**VISUALISASI BARPLOT SEDERHANA**




```python
# visualisasi Bar Plot Sederhana
plt.bar(data_total["Product Model"],data_total["Total"])
plt.xticks(rotation=90)
plt.show()
```

**VISUALISASI BAR PLOT DENGAN ELEMEN TAMBAHAN**

```python
# visualisasi bar plot dengan elemen tambahan

# menentukan dimensi
plt.figure(figsize=(10,5))

# plot data
plt.bar(data_total["Product Model"],data_total["Total"])
plt.xticks(size=10, rotation=90)
plt.yticks(size=10)
plt.xlabel("Model", size=12)
plt.ylabel("Total", size=12)
plt.title("Pengguna Samsung Berdasarkan Model", size=16)

#menampilkan data
plt.show()
```

**MENGATUR WARNA BAR**

```python
# menentukan dimensi
plt.figure(figsize=(10,5))

# plot data
plt.bar(data_total["Product Model"],data_total["Total"], color="skyblue")
plt.xticks(size=10, rotation=90)
plt.yticks(size=10)
plt.xlabel("Model", size=12)
plt.ylabel("Total", size=12)
plt.title("Pengguna Samsung Berdasarkan Model", size=16)

#menampilkan data
plt.show()
```

**MENGATUR WIDTH**

```python
# menentukan dimensi
plt.figure(figsize=(10,5))

# plot data
plt.bar(data_total["Product Model"],data_total["Total"], color="skyblue", width=0.65)
plt.xticks(size=10, rotation=90)
plt.yticks(size=10)
plt.xlabel("Model", size=12)
plt.ylabel("Total", size=12)
plt.title("Pengguna Samsung Berdasarkan Model", size=16)

#menampilkan data
plt.show()
```

**MENAMBAHKAN LABEL PADA BAR**

```python
# menentukan dimensi
plt.figure(figsize=(10,5))

data_total = data_total.sort_values("Total", ascending=False) #agar data urut dari yg terbesar
# plot data
bar_plot=plt.bar(data_total["Product Model"],data_total["Total"], color="skyblue", width=0.65)
plt.bar_label(bar_plot, fontsize=9, color="black")
plt.xticks(size=10, rotation=90)
plt.yticks(size=10)
plt.xlabel("Model", size=12)
plt.ylabel("Total", size=12)
plt.title("Pengguna Samsung Berdasarkan Model", size=16,)

#menampilkan data
plt.show()
```



> **HORIZONTAL BAR PLOT**



```python
plt.figure(figsize=(10,5))

# urutan dari besar ke kecil
data_total = data_total.sort_values("Total", ascending=False)

# plot data
bar_plot=plt.barh(data_total["Product Model"],data_total["Total"], color="skyblue", height=0.65)
plt.bar_label(bar_plot, fontsize=9, color="black")
plt.xticks(size=8)
plt.yticks(size=10)
plt.xlabel("Total", size=12)
plt.ylabel("Model", size=15)
plt.title("Pengguna Samsung Berdasarkan Model", size=16)
plt.gca().invert_yaxis() #agar nilai besar di atas
#menampilkan data
plt.show()
```



> **GROUPED BAR PLOT**


Untuk penggunaan grouped barplot menggunakan sample data penjualan samsung tahun 2023 dan 2024

```python
#filter data penjualan tahun 2023 dan 2024

data_sample = data[data["Year"].isin([2023, 2024])]
data_sample
```

```python
model=data_sample.groupby(["Product Model","Year"]).size()
data_model=pd.DataFrame(model).reset_index()
data_model=data_model.rename(columns={0:"Total"})
data_model
```

```python
# membuat list nilai unik kolom product model

product_model=list(data_model["Product Model"].unique())
product_model
```

```python
# membuat list nilai unik kolom year

year=list(data_model["Year"].unique())
year
```

```python
# filter dataframe product model dengan year

data_2023=data_model[data_model["Year"]==2023].sort_values(by="Product Model").reset_index(drop=True)
data_2024=data_model[data_model["Year"]==2024].sort_values(by="Product Model").reset_index(drop=True)

print(data_2023)
print(data_2024)
```

```python
# membuat grouped bar plot

# menentukan dimensi
plt.figure(figsize=(12, 6))

bar_width = 0.4
x = np.arange(len(data_2023["Product Model"]))

# plot data untuk 2023 dan 2024
bar1 = plt.bar(x, data_2023["Total"], bar_width, label='2023')
bar2 = plt.bar(x + bar_width, data_2024["Total"], bar_width, label="2024")

# menampilkan label dan judul
plt.xlabel("Model", size=12)
plt.ylabel("Total", size=12)
plt.title("Pengguna Samsung Berdasarkan Model (2023 vs 2024)", size=16)
plt.xticks(x + bar_width/2, data_2023["Product Model"], rotation=90, size=12)
plt.yticks(size=10)

# memunculkan legend
plt.legend(fontsize=12)

#menampilkan data
```

```python
# menambahkan label grouped bar plot
# ------------------------------------

# menentukan dimensi
plt.figure(figsize=(12, 6))

bar_width = 0.4
x = np.arange(len(data_2023["Product Model"]))

# plot data untuk 2023 dan 2024
bar1 = plt.bar(x, data_2023["Total"], bar_width, label='2023')
bar2 = plt.bar(x + bar_width, data_2024["Total"], bar_width, label="2024")

# menampilkan label dan judul
plt.xlabel("Model", size=12)
plt.ylabel("Total", size=12)
plt.title("Pengguna Samsung Berdasarkan Model (2023 vs 2024)", size=16)
plt.xticks(x + bar_width/2, data_2023["Product Model"], rotation=90, size=12)
plt.yticks(size=10)

# menampilkan label angka
plt.bar_label(bar1, padding=3)
plt.bar_label(bar2, padding=3)

# Amemunculkan legend
plt.legend(fontsize=12)

#menampilkan data
plt.tight_layout
```



> **STACKED BAR PLOT**



```python
# menentukan dimensi
plt.figure(figsize=(12, 6))

bar_width = 0.9
x = np.arange(len(data_2023["Product Model"]))

# plot data untuk 2023 dan 2024
bar1 = plt.bar(x, data_2023["Total"], bar_width, label='2023')
bar2 = plt.bar(x, data_2024["Total"], bar_width, label='2024', bottom=data_2023["Total"])

# menampilkan label dan judul
plt.xlabel("Model", size=12)
plt.ylabel("Total", size=12)
plt.title("Pengguna Samsung Berdasarkan Model (2023 vs 2024)", size=16)
plt.xticks(x, data_2023["Product Model"], rotation=90, size=12)
plt.yticks(size=10)

# menampilkan label angka
plt.bar_label(bar1, padding=3)
plt.bar_label(bar2, padding=3)

# Amemunculkan legend
plt.legend(fontsize=12)

#menampilkan data
plt.tight_layout
```
